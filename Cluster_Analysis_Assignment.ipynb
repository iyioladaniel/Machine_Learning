```
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.cluster import KMeans
import altair as alt
import random

file_url='https://raw.githubusercontent.com/PacktWorkshops/The-Data-Science-Workshop/master/Chapter05/DataSet/german.data-numeric'
df=pd.read_csv(file_url, header=None, sep='\s\s+', prefix='X')
print(df.describe().T)

#Only X3 and X9 have different variable and are numeric
X=df[['X3','X9']]
X

standard_scaler=StandardScaler()
standard_scaler.fit(X)
X_scaled=standard_scaler.transform(X)
```

```
#X['cluster1']=X_scaled.predict(X_scaled)

#getting perfect clusters using elbow (inertia)
clusters=pd.DataFrame()
clusters['cluster_range']=range(1,15)
inertia=[]

#create loop to to iterate over range
for k in clusters['cluster_range']:
  kmeans=KMeans(n_clusters=k, random_state=8).fit(X)
  inertia.append(kmeans.inertia_)

clusters['inertia']=inertia
print(clusters)

#plot line chart
alt.Chart(clusters).mark_line().encode(x='cluster_range',y='inertia').interactive()
```

```
x3_min=X['X3'].min()
x3_max=X['X3'].max()

x9_min=X['X9'].min()
x9_max=X['X9'].max()

print(x3_min,x3_max,x9_min,x9_max,sep='\n')

centroids=pd.DataFrame()
centroids['X3']=random.sample(range(x3_min,x3_max),5)
centroids['X9']=random.sample(range(x9_min,x9_max),5)
centroids['cluster']=centroids.index

chart2=alt.Chart(centroids).mark_circle(size=100).encode(x='X3',y='X9',color=alt.value('black'),
                                   tooltip=['cluster','X3','X9'])

kmeans=KMeans(random_state=100,n_clusters=5, init='k-means++',n_init=100)
kmeans.fit(X_scaled)
kmeans.transform(X_scaled)
df['cluster']=kmeans.predict(X_scaled)

chart1=alt.Chart(df).mark_circle().encode(x='X3',y='X9',color='cluster:N',
                                   tooltip=['cluster1','X3','X9']).interactive()

chart1+chart2
```
